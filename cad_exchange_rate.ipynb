```python
import urllib
import time
import numpy as np
import pandas as pd

import requests
from urllib.request import urlopen
import time

from bs4 import BeautifulSoup
```

```python
html = 'https://www.poundsterlinglive.com/bank-of-england-spot/historical-spot-exchange-rates/usd/USD-to-CAD-'

year = ['1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986']
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
for y in year:
    html2 = html + str(y)
    df = pd.read_html(html2)[-1].drop(['Unnamed: 2', 'Unnamed: 3'], axis=1)
    
    drop_list = []

    for j in range(len(df)):
        if df.iloc[j][0] in month:
            drop_list.append(j)
            
    df = df.drop(drop_list)
            
    for i in range(len(df)):
        df.iloc[i,0] = df.iloc[i,0].split(',')[1]
        df.iloc[i,0] = df.iloc[i,0].replace(' Dec ', '-12-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Nov ', '-11-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Oct ', '-10-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Sep ', '-09-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Aug ', '-08-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Jul ', '-07-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Jun ', '-06-')
        df.iloc[i,0] = df.iloc[i,0].replace(' May ', '-05-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Apr ', '-04-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Mar ', '-03-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Feb ', '-02-')
        df.iloc[i,0] = df.iloc[i,0].replace(' Jan ', '-01-')
        a = df.iloc[i,0].split('-')
        df.iloc[i,0] = a[2].strip() + '-' + a[1].strip() + '-' + a[0].strip()

        df.iloc[i,1] = df.iloc[i,1][8:14].strip()
        if 'C' in df.iloc[i,1]:
            df.iloc[i,1] = df.iloc[i,1].split(' ')[0]
        df.iloc[i,1] = float(df.iloc[i,1])
    
    df = df.sort_values(by='Date')
    
    df.to_csv('C:\\Users\\parks\\Desktop\\CAD_'+str(y)+'.csv')
```
